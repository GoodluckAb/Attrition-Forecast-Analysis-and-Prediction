```python
#encoding categorical variables
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
categorical_features = ['Attrition', 'Gender','OverTime']
for col in categorical_features:
    dataset[col] = encoder.fit_transform(dataset[col])
dataset[categorical_features].head()
```

```python
categorical_features_for_one_hot = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']
dataset = pd.get_dummies(dataset, columns=categorical_features_for_one_hot, drop_first=True)
dataset.head()
```

```python
#Over18 feature have only one value Yes, so  it is not effective and depends on the prediction so we can just remove it
dataset.drop(columns=['Over18'], inplace=True)
```

```python
# normalization of numberical data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical_features = dataset.select_dtypes(include=['int64','int32']).columns.tolist()
dataset[numerical_features]= scaler.fit_transform(dataset[numerical_features])
dataset.head()
```

```python
# convrting the bool values to numerical to avoid compatibility
bool_features = dataset.select_dtypes(include=['bool']).columns.tolist()
dataset[bool_features] = dataset[bool_features].astype(int)
dataset.dtypes
```

```python
#feature enginneering
#create a common variable employee satisfaction by including job satisfaction, environment satisfaction and relationship satisfaction
dataset['SatisfactionScore'] = (dataset['JobSatisfaction'] + dataset['EnvironmentSatisfaction'] + dataset['RelationshipSatisfaction']) / 3
```

```python
# employees with overtime and worklife balance are mutually depends
dataset['Overtime_WorkLifeBalance'] = dataset['OverTime'] * dataset['WorkLifeBalance']
```

```python
dataset['PerformanceRating'].values
```

```python
# identify the high perofrmeres with low salary who are supppose to affect the attrition
dataset['HighPerformerLowSalary'] = ((dataset['PerformanceRating'] >= 1) & (dataset['PercentSalaryHike'] < dataset['PercentSalaryHike'].mean())).astype(int)
```

```python
dataset.head()
```

```python
#model selecction
from sklearn.model_selection import train_test_split
X = dataset.drop('Attrition', axis=1)
y = dataset['Attrition']
```

```python
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import KFold
import numpy as np
# Train the model
#model = SVC(random_state=42)

#rf_model = RandomForestClassifier(random_state=42)
xgb_model = XGBClassifier(random_state=42)

#ensemble_model = VotingClassifier(estimators=[
    #('random_forest', rf_model),
    #('xgboost', xgb_model)
#], voting='hard')

kf = KFold(n_splits=5, shuffle=True, random_state=42)
confusion_matrices = []
classification_reports = []
accuracies = []

for fold, (train_index, test_index) in enumerate(kf.split(X), 1):
    print(f"Fold {fold}:")
    X_train_fold, X_test_fold = X.iloc[train_index], X.iloc[test_index]
    y_train_fold, y_test_fold = y.iloc[train_index], y.iloc[test_index]
    xgb_model.fit(X_train_fold, y_train_fold)
    y_pred_fold = xgb_model.predict(X_test_fold)
    conf_matrix = confusion_matrix(y_test_fold, y_pred_fold)
    class_report = classification_report(y_test_fold, y_pred_fold, output_dict=True)
    accuracy = np.mean(y_pred_fold == y_test_fold)
    
    confusion_matrices.append(conf_matrix)
    classification_reports.append(class_report)
    accuracies.append(accuracy)
    print("Confusion Matrix:")
    print(conf_matrix)
    print("\nClassification Report:")
    print(classification_report(y_test_fold, y_pred_fold))
    print(f"Accuracy for fold {fold}: {accuracy}")
    print('-' * 50)

# Print overall results
print(f"\nOverall Accuracy: {np.mean(accuracies)}")
```
